```python
columns=['sepal_length','sepal_width','petal_length','petal_width','species']
```

```python
iris=pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',names=columns)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.boxplot(x=iris.species,y=iris.sepal_length)
plt.subplot(2,2,2)
sns.boxplot(x=iris.species,y=iris.sepal_width)
plt.subplot(2,2,3)
sns.boxplot(x=iris.species,y=iris.petal_length)
plt.subplot(2,2,4)
sns.boxplot(x=iris.species,y=iris.petal_width)
plt.show()
```

```python
Bivariate analysis of numerical-categorical
```

```python
plt.figure(figsize=(13,13))
plt.subplot(2,2,1)
sns.barplot(x=iris.species,y=iris.sepal_length)
plt.subplot(2,2,2)
sns.barplot(x=iris.species,y=iris.sepal_width)
plt.subplot(2,2,3)
sns.barplot(x=iris.species,y=iris.petal_length)
plt.subplot(2,2,4)
sns.barplot(x=iris.species,y=iris.petal_width)
plt.show()
```

```python
Bi-variate analysis Numerical-Numerical
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.distplot(iris['petal_length'])
plt.subplot(2,2,2)
sns.distplot(iris['petal_width'])
plt.subplot(2,2,3)
sns.distplot(iris['sepal_length'])
plt.subplot(2,2,4)
sns.distplot(iris['sepal_width'])
plt.show()
```

```python
sns.pairplot(iris,kind='scatter',hue='species')
plt.show()
```

```python
iris.groupby('species').mean().plot.bar()
```

```python
Multivariate analysis
```

```python
plt.figure(figsize=(10,8))
corr=iris.corr()
sns.heatmap(corr,cmap='RdYlGn',annot=True,vmin=-1.0,center=0)
plt.show()
```
